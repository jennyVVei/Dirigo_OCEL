```python
import pandas as pd
import ast
```

```python
events_df = pd.read_csv("./Dirigo/OCEL_Events.csv")
#events_df
```

```python
acel_events_df = events_df.copy()
```

```python
acel_events_df["Objects"] = None
acel_events_df["Relations"] = None
acel_events_df["ObjectChanges"] = None
acel_events_df["RelationChanges"] = None
#acel_events_df
```

```python
e2o_df = pd.read_csv("./Dirigo/OCEL_E2O.csv")
e2o_df
```

```python
objects_by_event = e2o_df.groupby("Event_id")["Object_id"].apply(list).to_dict()
objects_by_event
```

```python
acel_events_df["Objects"] = acel_events_df["Event_id"].map(objects_by_event)
acel_events_df
# -- 1500 rows Ã— 7 columns
```

```python
# Map Timstamp to Event ID
acel_events_df["Timestamp"] = acel_events_df["Timestamp"].astype(str)
event_time_map = dict(zip(acel_events_df["Event_id"], acel_events_df["Timestamp"]))
timestamp_to_event = {v: k for k, v in event_time_map.items()}
timestamp_to_event
```

```python
o2o_df = pd.read_csv("./Dirigo/OCEL_O2O.csv")
# o2o_df
# -- 500 rows Ã— 4 columns
```

```python
o2o_df["Timestamp"] = o2o_df["Timestamp"].astype(str)
```

```python
o2o_df["CorrectedSource"] = o2o_df["Target_object_id"]
o2o_df["CorrectedTarget"] = o2o_df["Source_object_id"]
o2o_df["RelationId"] = o2o_df["CorrectedSource"] + "_to_" + o2o_df["CorrectedTarget"]
```

```python
o2o_df["Event_id"] = o2o_df["Timestamp"].map(timestamp_to_event)
o2o_df
```

```python
relations_table = o2o_df[[
    "RelationId", "O2O_qualifier", "CorrectedSource", "CorrectedTarget", "Timestamp"
]].copy()
#relations_table
```

```python
relations_table.rename(columns={
    "O2O_qualifier": "Type",
    "CorrectedSource": "Source",
    "CorrectedTarget": "Target"
}, inplace=True)
relations_table["ChangeStatus"] = "addedTarget"
```

```python
relations_table["Cardinality"] = "unknown"
cargo_to_pcp_mask = (
    relations_table["Source"].str.startswith("Cr") &
    relations_table["Target"].str.startswith("Pcp")
)
relations_table.loc[cargo_to_pcp_mask, "Cardinality"] = "one2Many"
pcp_to_truck_mask = (
    relations_table["Source"].str.startswith("Pcp") &
    relations_table["Target"].str.startswith("tr")
)
relations_table.loc[pcp_to_truck_mask, "Cardinality"] = "one2Many"
relations_table
```

```python
event_timestamp_map = dict(zip(acel_events_df["Event_id"], acel_events_df["Timestamp"].astype(str)))
event_timestamp_map
```

```python
filtered_relations_by_event = {}

for event_id, timestamp in event_timestamp_map.items():
    event_objs = set(objects_by_event.get(event_id, []))

    candidate_relations = relations_table[relations_table["Timestamp"] == timestamp]

    matched = candidate_relations[
        candidate_relations["Target"].isin(event_objs)
    ]

    relation_ids = matched["RelationId"].tolist()
    if relation_ids:
        filtered_relations_by_event[event_id] = relation_ids

acel_events_df["Relations"] = acel_events_df["Event_id"].map(filtered_relations_by_event)
```

```python
# acel_events_df.to_csv("acel_events_relations.csv",index=False)
```

```python
summary_relations = relations_table[["RelationId", "Type", "Source", "Cardinality"]]
summary_relations
```

```python
def revise_type(row):
    source = row["Source"]
    target = row["RelationId"].split("_to_")[1]

    if source.startswith("Cr") and target.startswith("Pcp"):
        return "is lodged to pick up"
    elif source.startswith("Pcp") and target.startswith("tr"):
        original = row["Type"].lower()
        if "assigned" in original:
            return "is assigned to"
        elif "dropped" in original:
            return "is dropped from"
    return row["Type"]
```

```python
summary_relations["Type"] = summary_relations.apply(revise_type, axis=1)
summary_relations
```

```python
summary_relations.to_csv('./ACEL/ACEL_Relations.csv',index=False)
```

```python
relation_changes_by_event_time = (
    relations_table
    .groupby("Timestamp")
    .apply(lambda df: [
        {
            "RelationId": row["RelationId"],
            "Target": row["Target"],
            "ChangeStatus": row["ChangeStatus"]
        }
        for _, row in df.iterrows()
    ])
    .to_dict()
)
relation_changes_by_event_time
```

```python
filtered_relation_changes_by_event = {}
for event_id, timestamp in event_timestamp_map.items():
    all_changes = relation_changes_by_event_time.get(timestamp, [])
    event_targets = set(objects_by_event.get(event_id, []))

    filtered = [change for change in all_changes if change["Target"] in event_targets]

    if filtered:
        filtered_relation_changes_by_event[event_id] = filtered
```

```python
filtered_relation_changes_by_event
```

```python
acel_events_df["RelationChanges"] = acel_events_df["Event_id"].map(filtered_relation_changes_by_event)
```

```python
event_activity_map = dict(zip(acel_events_df["Event_id"], acel_events_df["Activity"]))
event_activity_map
```

```python
def fix_relation_change(row):
    changes = filtered_relation_changes_by_event.get(row["Event_id"])
    if isinstance(changes, list) and row["Activity"] == "Evaluate the Truck Exit":
        for c in changes:
            c["ChangeStatus"] = "deletedTarget"
    return changes
```

```python
acel_events_df["RelationChanges"] = acel_events_df.apply(fix_relation_change, axis=1)
acel_events_df
```

```python
cargo_df = pd.read_csv("./Dirigo/OCEL_Cargo.csv")
truck_df = pd.read_csv("./Dirigo/OCEL_Truck.csv")
pickup_plan_df = pd.read_csv("./Dirigo/OCEL_PickupPlan.csv")
```

```python
# truck_df.rename(columns={
#     "PickupPlanID": "Pickup Plan ID",
#     "CargoID": "Cargo ID"
# }, inplace=True)
```

```python
cargo_df["Type"] = "Cargo"
truck_df["Type"] = "Truck"
pickup_plan_df["Type"] = "PickupPlan"
```

```python
cargo_truck_df = pd.concat([cargo_df, truck_df], ignore_index=True)
dynamic_attr_df = cargo_truck_df[cargo_truck_df["Ocel_changed_field"].notnull()].copy()
```

```python
dynamic_attr_df 
```

```python
cargotr_dynamic_records = []

for _, row in dynamic_attr_df.iterrows():
    attr = row["Ocel_changed_field"]
    
    new_value = row.get(attr, None)
    cargotr_dynamic_records.append({
        "Object_id": row["Object_id"],
        "Attribute": attr,
        "NewValue": new_value,
        "Timestamp": row["Timestamp"],
        "Type": row["Type"]
    })
```

```python
pickup_cargotr_dynamic = pd.DataFrame(cargotr_dynamic_records)
pickup_cargotr_dynamic
```

```python
cargotr_dynamic_records
```

```python
pickup_dynamic_records = []
meta_columns = {"Object_id", "Timestamp", "Ocel_changed_field", "Type"}

# Extract dynamic attributes from PickupPlan table
for _, row in pickup_plan_df.iterrows():
    for col in pickup_plan_df.columns:
        if col not in meta_columns:
            value = row[col]
            if pd.notnull(value):
                pickup_dynamic_records.append({
                    "Object_id": row["Object_id"],
                    "Attribute": col,
                    "NewValue": value,
                    "Timestamp": row["Timestamp"],
                    "Type": row["Type"]
                })

# # Convert to DataFrame
pickup_dynamic_df = pd.DataFrame(pickup_dynamic_records)
pickup_dynamic_df
```

```python
object_changes_df = pd.DataFrame(cargotr_dynamic_records + pickup_dynamic_records)
object_changes_df
```

```python
timestamp_object_to_activity = {}

for event_id, obj_list in objects_by_event.items():
    ts = event_time_map.get(event_id)
    act = event_activity_map.get(event_id)
    for obj in obj_list:
        timestamp_object_to_activity[(ts, obj)] = act

object_changes_df2 = object_changes_df.copy()

object_changes_df2["Activity"] = object_changes_df2.apply(
    lambda row: timestamp_object_to_activity.get((str(row["Timestamp"]), row["Object_id"]), "Unknown"),
    axis=1
)
```

```python
object_changes_df2   
```

```python
qualifier_to_lifecycle = {
    "Cargo being scheduled for pickup": "scheduled",
    "Pickup plan being lodged": "lodged",
    "Truck being assigned for pickup": "assigned",
    "Truck being evaluated for exit": "released",
    "Truck being loaded with cargo": "loading",
    "Empty truck getting weighed": "weigh_empty",
    "Loaded truck getting weighed": "weigh_loaded",
    "Truck being recorded on tally sheet": "tally_recorded",
    "Truck receiving weighing ticket": "ticket_issued"
}
qualifier_to_lifecycle
```

```python
lifecycle_records = []
for _, row in e2o_df.iterrows():
    qualifier = row["E2O_qualifier"]
    lifecycle_value = qualifier_to_lifecycle.get(qualifier)
    event_id = row["Event_id"]

    if lifecycle_value:
        lifecycle_records.append({
            "Object_id": row["Object_id"],
            "Attribute": "lifecycle",
            "NewValue": lifecycle_value,
            "Timestamp": event_time_map.get(event_id),
            "Type": None,
            "Activity": event_activity_map.get(event_id)
        })

lifecycle_df = pd.DataFrame(lifecycle_records)
lifecycle_df
```

```python
object_type_map = {}
for df, obj_type in [(cargo_df, "Cargo"), (truck_df, "Truck"), (pickup_plan_df, "PickupPlan")]:
    object_type_map.update({obj_id: obj_type for obj_id in df["Object_id"]})
lifecycle_df["Type"] = lifecycle_df["Object_id"].map(object_type_map)
lifecycle_df
```

```python
final_object_changes_df = pd.concat([object_changes_df2, lifecycle_df], ignore_index=True)
final_object_changes_df
```

```python
object_changes_index = {}

for _, row in final_object_changes_df.iterrows():
    timestamp = str(row["Timestamp"])
    object_id = row["Object_id"]
    activity = row["Activity"]
    key = (timestamp, object_id, activity)

    if key not in object_changes_index:
        object_changes_index[key] = []

    object_changes_index[key].append({
        "ObjectId": object_id,
        "Attribute": row["Attribute"],
        "NewValue": row["NewValue"]
    })
```

```python
object_changes_index
```

```python
filtered_object_changes_by_event = {}

for idx, row in acel_events_df.iterrows():
    event_id = row["Event_id"]
    timestamp = str(row["Timestamp"])
    activity = row["Activity"]

    try:
        objects = ast.literal_eval(row["Objects"]) if isinstance(row["Objects"], str) else row["Objects"]
    except:
        objects = []

    changes = []
    for obj in objects:
        key = (timestamp, obj, activity)
        matches = object_changes_index.get(key, [])
        if matches:
            changes.extend(matches)

    if changes:
        filtered_object_changes_by_event[event_id] = changes
```

```python
filtered_object_changes_by_event
```

```python
acel_events_df["ObjectChanges"] = acel_events_df["Event_id"].map(filtered_object_changes_by_event)
```

```python
def extract_static_objects_strict(cargo_truck_df):
    # Step 1: All attributes that are *ever* marked as dynamic
    dynamic_attrs = set(cargo_truck_df["Ocel_changed_field"].dropna().unique())

    # Step 2: Static attributes = columns not listed as dynamic, and not metadata
    exclude_cols = dynamic_attrs.union({"Ocel_changed_field", "Timestamp"})
    static_cols = [col for col in cargo_truck_df.columns if col not in exclude_cols]

    # Step 3: Get the first version of each object based on Object_id
    static_objects_df = cargo_truck_df[static_cols].drop_duplicates(subset=["Object_id"]).copy()

    # Step 4: Rename column for ACEL
    static_objects_df = static_objects_df.rename(columns={"Object_id": "ObjectId"})

    return static_objects_df
```

```python
acel_objects_df = extract_static_objects_strict(cargo_truck_df)
acel_objects_df
```

```python
acel_objects_df.rename(columns={"Type": "Object Type"}, inplace=True)
```
